```python
df['lower_case'] = df['ocr_text'].str.lower()
```

```python
df[['id', 'ocr_text', 'lower_case']].head()
```

```python
df['alpha_text'] = df.lower_case.str.replace('[^a-zA-Z]', ' ')
```

```python
df[['alpha_text']].head()
```

```python
# this came up in the workshop
# https://pandas.pydata.org/docs/reference/api/pandas.Series.str.replace.html
# if you're going to use a regex, you need to set regex=True, it's False by default
# thanks for the suggestions from the class on this one!
```

```python
# this replaces every non-alpha character with a space. 
# but this creates a lot of whitespace...
df["alpha_text"] = df.lower_case.str.replace("[^a-z]", ' ', regex=True)
```

```python
df["alpha_text"].head()
```

```python
# 
# (yes, I also used stackoverflow for this one:
# https://stackoverflow.com/questions/2077897/substitute-multiple-whitespace-with-single-whitespace-in-python
# using the regular expression approach

import re
df["alpha_text"] = df["alpha_text"].str.replace(' +',' ', regex=True)
```
